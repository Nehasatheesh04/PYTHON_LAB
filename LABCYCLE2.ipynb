```python
#1 
def rabbit_pairs(n):
    pairs = [1, 1]
    for i in range(2, n):
        pairs.append(pairs[i-1] + pairs[i-2])
    return pairs

n = int(input("Enter the number of months: "))
pairs = rabbit_pairs(n)

print("Month\tPairs")
for i in range(n):
    print(f"{i+1}\t{pairs[i]}")
```

```python
#2 
# Read the string and convert it to a list of integers
num_str = input("Enter numbers separated by spaces: ")
num_list = [int(num) for num in num_str.split()]

# 1. Rotate elements in a list by 'k' positions to the right
k = int(input("Enter the number of positions to rotate: "))
rotated_list = num_list[-k:] + num_list[:-k]
print("Rotated list:", rotated_list)

# 2. Convert the list into a tuple using list comprehension
tuple_list = tuple([num for num in num_list])
print("Tuple list:", tuple_list)

# 3. Remove duplicates from the tuple and convert it to a list again
unique_list = list(set(tuple_list))
print("Unique list:", unique_list)

# 4. Create another list by applying the function f(x) = x^2 - x to each element
def f(x):
    return x**2 - x

f_list = [f(num) for num in unique_list]
print("f(x) list:", f_list)

# 5. Merge and sort the two lists into a single sorted list
merged_list = sorted(unique_list + f_list)
print("Merged and sorted list:", merged_list)
```

```python
#5
import math

class ThreeDShapes:
    def printVolume(self):
        pass

    def printArea(self):
        pass

class Cylinder(ThreeDShapes):
    def __init__(self):
        self.radius = float(input("Enter the radius of the cylinder: "))
        self.height = float(input("Enter the height of the cylinder: "))

    def printVolume(self):
        volume = math.pi * self.radius * self.radius * self.height
        print("Volume of the cylinder: ", volume)

    def printArea(self):
        area = 2 * math.pi * self.radius * (self.radius + self.height)
        print("Surface area of the cylinder: ", area)

class Sphere(ThreeDShapes):
    def __init__(self):
        self.radius = float(input("Enter the radius of the sphere: "))

    def printVolume(self):
        volume = (4/3) * math.pi * self.radius * self.radius * self.radius
        print("Volume of the sphere: ", volume)

    def printArea(self):
        area = 4 * math.pi * self.radius * self.radius
        print("Surface area of the sphere: ", area)

c = Cylinder()
c.printVolume()
c.printArea()

s = Sphere()
s.printVolume()
s.printArea()
```

```python
#4 
import random

class Box:
    def __init__(self, *args):
        if len(args) == 1:
            self.length = self.breadth = self.height = args[0]
        elif len(args) == 2:
            self.length, self.breadth = args
            self.height = self.length
        elif len(args) == 3:
            self.length, self.breadth, self.height = args
        self.area = 2 * (self.length * self.breadth + self.breadth * self.height + self.height * self.length)
        self.volume = self.length * self.breadth * self.height

# Creating a list of N boxes with random measurements
N = 5
boxes = [Box(random.randint(1, 10), random.randint(1, 10), random.randint(1, 10)) for i in range(N)]

# Finding the box with maximum volume : area ratio
max_ratio = 0
max_box = None
for box in boxes:
    ratio = box.volume / box.area
    if ratio > max_ratio:
        max_ratio = ratio
        max_box = box

# Printing the details of the box with maximum volume : area ratio
print("Box with maximum volume : area ratio")
print("Length:", max_box.length)
print("Breadth:", max_box.breadth)
print("Height:", max_box.height)
print("Area:", max_box.area)
print("Volume:", max_box.volume)
```

```python
#3 
import json
def create_list(way):
  file1=open(way,"r")
  result=file1.read()
  file1.close()
  return result

def list_of_dict(list1):
  dummy=json.loads(list1)
  return (dummy)

def setosa(dlist):
  print("Details of all Setosa Flowers:")
  for i in dlist:
    if i["species"]=="setosa":
      print(i)


def max_min(dlist):
  species=[]
  
  for i in dlist:
    species.append(i["species"])

  species=list(set(species))
  sepal_Area=[]
  petal_Area=[]

  for i in species:
    for j in dlist:
      if j["species"]==i:
        sepal_Area.append(j["sepalLength"]*j["sepalWidth"])
        petal_Area.append(j["petalLength"]*j["petalWidth"])
    
    print("Minimum petal area of",i,"=",round(min(petal_Area),2))
    print("Maximum sepal area of",i,"=",round(max(sepal_Area),2))

def total_area(dlist):
  for i in dlist:
    p_area=i["petalLength"]*i["petalWidth"]
    s_area=i["sepalLength"]*i["sepalWidth"]
    total_area=p_area+s_area
    i.update({"TotalArea":round(total_area,2)})

  sortedlist=sorted(dlist, key=lambda i:i["TotalArea"])
  for i in sortedlist:
    print(i)


way= "/content/drive/MyDrive/iris.json"
list1=create_list(way)
print("List with each line of file as elements:")
print(list1,"\n")

dlist=list_of_dict(list1)
print("List of Dictionary objects:")
print(dlist,"\n")
setosa(dlist)
print("\n")
max_min(dlist)
print("\n")
print("Sorted according to Total Area")
total_area(dlist)
```
